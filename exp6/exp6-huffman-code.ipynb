```python
freqs = get_freqs(image)
```

```python
coding_scheme = {}

def printNodes(node, val=''): 
    global node_store
    newVal = val + str(node.huff) 

    if(node.left): 
        printNodes(node.left, newVal) 
    if(node.right): 
        printNodes(node.right, newVal) 

    if(not node.left and not node.right): 
        coding_scheme[node.symbol] = newVal
    return
```

```python
nodes = huffman(freqs)
```
