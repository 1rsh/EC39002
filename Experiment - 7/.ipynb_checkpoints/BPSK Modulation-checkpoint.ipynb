**Generating Huffman Code**

```python
transmitted = huffmanenco(image)
```

```python
def bpsk_modulation(bits, samples_per_symbol):
    phase_states = [0, np.pi]
    phase_sequence = [phase_states[int(b)] for b in bits]
    t = np.arange(0,len(phase_sequence))/10e3
    fc = 1e3
    carrier_signal = np.sin(2*np.pi*fc*t+phase_signal)
    return carrier_signal
```

```python
sins = bpsk_modulation(transmitted, 1)
```

```python
plt.plot(sins[:300])
plt.show()
```

```python
recovered_img = decoder(transmitted, image.shape[0], image.shape[1])

plt.imshow(recovered_img, cmap = "gray")
```
