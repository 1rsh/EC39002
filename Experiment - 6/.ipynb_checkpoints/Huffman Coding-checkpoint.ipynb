**Reading in the image**

```python
FILEPATH = "bmw1.png"
image = read_img(FILEPATH)
IM_SHAPE_X, IM_SHAPE_Y = image.shape[:2]
```

```python
plt.imshow(image, cmap="gray")
```

**Getting Frequencies of all Pixel Values**

```python
freqs = get_freqs(image)
```

```python
coding_scheme = {}

def printNodes(node, val=''): 
    global node_store
    newVal = val + str(node.huff) 

    if(node.left): 
        printNodes(node.left, newVal) 
    if(node.right): 
        printNodes(node.right, newVal) 

    if(not node.left and not node.right): 
        coding_scheme[node.symbol] = newVal
    return
```

```python
nodes = huffman(freqs)
```

```python
printNodes(nodes[0])
```

**Generating Huffman Code**

```python
transmitted = transmitter(image, coding_scheme, IM_SHAPE_X, IM_SHAPE_Y)
```

```python
print(f"Length of Huffman Code: {len(transmitted)}")
```

```python
recovered_img = decoder(transmitted, IM_SHAPE_X, IM_SHAPE_Y, coding_scheme)

plt.imshow(recovered_img, cmap = "gray")

print(f"# of Bits if transmitted without coding: {IM_SHAPE_X * IM_SHAPE_Y * 8}")
print(f"# of Bits if transmitted with Huffman coding: {len(transmitted)}")
print(f"Percentage Decrease: {1 - len(transmitted)/(IM_SHAPE_X * IM_SHAPE_Y * 8) :.4f} | Average Length of Code: {len(transmitted)/(IM_SHAPE_X * IM_SHAPE_Y):.4f}")
```
